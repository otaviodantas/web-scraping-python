```python
from urllib.request import Request, urlopen
from urllib.error import URLError, HTTPError
from my_options import URL_INPUT, HEADER
from url_handling import handling_url
from dataframe import captch_data
from bs4 import BeautifulSoup
import pandas as pd
url_input = URL_INPUT
header = HEADER

try:
    request = Request(url_input, headers=header)
    response = urlopen(request)

except HTTPError as e:
    print(e.status, e.reason)

except URLError as e:
    print(e.reason)
```

```python
html_clean = handling_url(response)
soup = BeautifulSoup(html_clean, 'html.parser')

result_anuncio = soup.find(
    'div', class_='sc-1fcmfeb-0 WQhDk').findAll('div', class_="fnmrjs-1 ddDPXY")

cards = []
card = {}
```

```python
for item in result_anuncio:
    card['values'] = item.find('p', 'fnmrjs-16 jqSHIm').getText()
    card['title'] = item.find('div', 'fnmrjs-8 kRlFBv').getText()
    card['location'] = item.find('p', 'fnmrjs-13 hdwqVC').getText()
    card['category'] = item.find('p', 'fnmrjs-14 iIprpQ').getText()
    print(card)
    cards.append(card)

        
dataset = pd.DataFrame(cards)
```
